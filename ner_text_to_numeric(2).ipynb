Convert TRAMER to numeric value

```python
df1['TRAMER']
```

```python
import re

tramer_ozellik_numeric= []

for i, text in enumerate(df1['TRAMER'], start=1):
  if text!='[]':
    pattern=r"(\d+)\s*(\d+)"
    matches=re.findall(pattern, text)
    tramer_ozellik_numeric.append(matches)
  else:
    sayı=0
    tramer_ozellik_numeric.append(sayı)
```

```python
#tramer_ozellik_numeric=[]
```

```python
result = []

for tramer in tramer_ozellik_numeric:
    if tramer:
      concatenated = ''.join(tramer[0])  # Concatenate inner tuple elements
      result.append(int(concatenated))  # Convert concatenated string to integer
    else:
      sayı=0
      result.append(sayı)
print(result)
```

```python
len(tramer_ozellik_numeric)
```

```python
# Add the new column using loc
df1.loc[:, "TRAMER NUMERİK"] = result
```

Convert Boya to numeric

```python
df1['BOYA']
```

```python
boya_sayısı_numeric= []

for i, text in enumerate(df1['BOYA'], start=1):
  if text=='[]':
    sayı= 0
  elif "," in text:
        sayı = text.count(",") + 1
  elif  any(char.isdigit() for char in text):
      sayı = int(''.join(filter(str.isdigit, text)))
      if  sayı >=  10 and sayı < 100:
          sayı = int(sayı/10)
      elif sayı > 100:
       sayı = int(str(sayı)[0])
  else:
    sayı =1
  boya_sayısı_numeric.append(sayı)
```

```python
df1.loc[:, "BOYA SAYISI NUMERİC"] = boya_sayısı_numeric
```

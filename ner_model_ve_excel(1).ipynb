This code converts the training data into a spaCy compatible format and saves this data to disk. In this way, we can easily access data during the training process.

```python
import pandas as pd
import os
from tqdm import tqdm
import spacy
from spacy.tokens import DocBin

#nlp = spacy.blank("en") # load a new spacy model
nlp = spacy.load("tr_core_news_lg") # load other spacy model

db = DocBin() # create a DocBin object

for text, annot in tqdm(train_data): # data in previous format
    doc = nlp.make_doc(text) # create doc object from text
    ents = []
    for start, end, label in annot["entities"]: # add character indexes
        span = doc.char_span(start, end, label=label, alignment_mode="contract")
        if span is None:
            print("Skipping entity")
        else:
            ents.append(span)
    doc.ents = ents # label the text with the ents
    db.add(doc)

#os.chdir(r'XXXX\XXXXX')
db.to_disk("./train.spacy") # save the docbin object
```

```python
#define the columns as array
degisen = []
tramer = []
yedekanahtar = []
boya = []
bakim = []
degisenyok = []
iyi = []
kotu = []
bilgiyok= []
```

```python
nlp1 = spacy.load("/content/output/model-best") #load the best model
```

```python
#this block adds the entities to relevant arrays
for i, text in enumerate(df['Processed Açıklama Metni'], start=1):
    degisen_ = []
    tramer_ = []
    yedekanahtar_ = []
    boya_ = []
    bakim_ = []
    degisenyok_ = []
    iyi_ = []
    kotu_ = []
    bilgiyok_ = []

    doc = nlp1(text)
    for token in doc.ents:
        if(token.label_ == "DEGISEN"):
            degisen_.append(token.text)
        elif(token.label_ == "TRAMER"):
            tramer_.append(token.text)
        elif(token.label_ == "YEDEK ANAHTAR"):
            yedekanahtar_.append(token.text)
        elif(token.label_ == "BOYA"):
            boya_.append(token.text)
        elif(token.label_ == "BAKIM DURUMU"):
            bakim_.append(token.text)
        elif(token.label_ == "DEGISEN YOK"):
            degisenyok_.append(token.text)
        elif(token.label_ == "IYI OZELLIK"):
            iyi_.append(token.text)
        elif(token.label_ == "KOTU OZELLIK"):
            kotu_.append(token.text)
        elif(token.label_ == "BILGI YOK"):
            bilgiyok_.append(token.text)


    degisen.append(degisen_)
    tramer.append(tramer_)
    yedekanahtar.append(yedekanahtar_)
    boya.append(boya_)
    bakim.append(bakim_)
    degisenyok.append(degisenyok_)
    iyi.append(iyi_)
    kotu.append(kotu_)
    bilgiyok.append(bilgiyok_)


    # Track the iteration count
    print(f"Iteration {i} completed.")
```

```python
# Add columns to dataframe

df["DEGISEN"] = degisen
df["TRAMER"] = tramer
df["YEDEK ANAHTAR"] = yedekanahtar
df["BOYA"] = boya
df["BAKIM DURUMU"] = bakim
df["DEGISEN YOK"] = degisenyok
df["IYI OZELLIK"] = iyi
df["KOTU OZELLIK"] = kotu
df["BILGI YOK"] = bilgiyok


# Export the daga as xlsx
df.to_excel("project_dataset_nerli2.xlsx", index=False)
```

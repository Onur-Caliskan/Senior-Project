```python
import numpy as np
import pandas as pd
import re
```

```python
df1 = pd.read_excel("data.xlsx")
```

```python
def clean_non_numeric(text):#This code gets numerical characters in the data
    # if the data type is not string, convert it to string
    if not isinstance(text, str):
        text = str(text)

    # get numerical characters
    return ''.join(filter(str.isdigit, text))
```

```python
def ortalama_hesapla(sayi):#this code gets mean of two number to get correct data
    # if the length of number is 8 aplly the process
    if len(str(sayi)) == 8:
        # Separate the first 4 digits and the next 4 digits and take the average
        ilk_dort_hane = int(str(sayi)[:4])
        sonraki_dort_hane = int(str(sayi)[4:])
        return (ilk_dort_hane + sonraki_dort_hane) / 2
    else:
        # for the other situations return actual number
        return sayi
```

```python
def ortalama_hesapla2(sayi):#this code gets mean of two number to get correct data
    # if the length of number is 6 aplly the process
    if len(str(sayi)) == 6:
        # Separate the first 3 digits and the next 3 digits and take the average
        ilk_dort_hane = int(str(sayi)[:3])
        sonraki_dort_hane = int(str(sayi)[3:])
        return (ilk_dort_hane + sonraki_dort_hane) / 2
    else:
        # for the other situations return actual number
        return sayi
```

```python
def ikinci_yariyi_al(sayi):#this function gets second half to get correct data
    sayi_str = str(sayi)
    # if the length of number greater than 10 apply the process
    if len(sayi_str) > 10:
        ikinci_yari = sayi_str[len(sayi_str)//2:]  # get second half
        return int(ikinci_yari)
    else:
        return sayi  # For the other situations return actual number
```

```python
#Apply the functions for Fiyat Column
df1['Fiyat'] = df1['Fiyat'].apply(clean_non_numeric)
df1['Fiyat'] = df1['Fiyat'].apply(ikinci_yariyi_al)
```

```python
#Apply the functions for İlan No Column
df1['Fiyat'] = df1['Fiyat'].apply(clean_non_numeric)
df1['Fiyat'] = df1['Fiyat'].apply(ikinci_yariyi_al)

df1['İlan No'] = df1['İlan No'].apply(clean_non_numeric)



df1['İlan No'] = df1['İlan No'].astype(int)
```

```python
#Apply the functions for Motor Gücü Column
df1['Motor Gücü'] = df1['Motor Gücü'].apply(clean_non_numeric)

df1['Motor Gücü'] = df1['Motor Gücü']

df1['Motor Gücü'] = df1['Motor Gücü'].apply(ortalama_hesapla2)

df1['Motor Gücü'] = df1['Motor Gücü']
```

```python
#Apply the functions for Motor Hacmi Column
df1['Motor Hacmi'] = df1['Motor Hacmi'].apply(clean_non_numeric)

df1['Motor Hacmi'] = df1['Motor Hacmi']

df1['Motor Hacmi'] = df1['Motor Hacmi'].apply(ortalama_hesapla)

df1['Motor Hacmi'] = df1['Motor Hacmi']
```

```python
#Apply the functions for Km Column
df1['Km'] = df1['Km'].apply(clean_non_numeric)
df1['Km'] = df1['Km'].apply(ikinci_yariyi_al)
```

```python
#Apply the functions for Yakıt Deposu Column
df1['Yakıt Deposu'] = df1['Yakıt Deposu'].apply(clean_non_numeric)
df1['Yakıt Deposu'] = df1['Yakıt Deposu'].apply(ikinci_yariyi_al)
```

```python
#Apply the functions for Ort. Yakıt Tüketimi Column
df1['Ort. Yakıt Tüketimi'] = df1['Ort. Yakıt Tüketimi'].str.extract(r'(\d+,\d+|\d+)')
df1['Ort. Yakıt Tüketimi'] = df1['Ort. Yakıt Tüketimi'].apply(lambda x: float(x.split(' ')[0].replace(',', '.')) if pd.notnull(x) else np.nan)

print(df1)
```

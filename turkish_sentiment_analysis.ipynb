```python
  import pandas as pd
```

```python
data = pd.read_excel("Toplu_Duplicatesiz_Nansız_2.xlsx")#load the dataset
```

```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer, pipeline
tokenizer = AutoTokenizer.from_pretrained("savasy/bert-base-turkish-sentiment-cased")#load the tokenizer
model = AutoModelForSequenceClassification.from_pretrained("savasy/bert-base-turkish-sentiment-cased")#load the model
```

```python
def sentiment(aciklama_metni):#this function gets sentiment and sentiment score of the text
  duygu_analizi_modeli = pipeline("sentiment-analysis", tokenizer=tokenizer, model=model)
  sonuc= duygu_analizi_modeli(aciklama_metni)
  print(sonuc)
  # print the result
  for duygu in sonuc:
      etiket, skor = duygu['label'], duygu['score']
      print(f"{etiket}: {skor}")

  # get the label that has highest score
  en_yuksek_skorlu_etiket = max(sonuc, key=lambda x: x['score'])['label']
  print(f"En yüksek skorlu duygu: {en_yuksek_skorlu_etiket}")
```

```python
# this function truncates long texts to fit the model's maximum sequence length
def truncate_text(text):
    max_seq_length = 512
    if len(tokenizer.encode(text)) > max_seq_length:
        return text[:max_seq_length]
    return text
```

```python
#this code block gets the sentiment and score of the dataframe's text and appends them to list
sentiments = []
sent_scores = []
for i, text in enumerate(data['Processed Açıklama Metni'], start=1):
    truncated_text = truncate_text(text)
    result = duygu_analizi_modeli(truncated_text)
    label = result[0]['label']
    sentiments.append(label)

    score = result[0]['score']
    sent_scores.append(score)

    # Track the iteration count
    print(f"Iteration {i} completed.")
```

```python
# Add sentiment labels to the DataFrame
data['sentiment'] = sentiments
data['sent-scores'] = sent_scores

# export dataframe as xlsx
data.to_excel("sent_score_tokenlimodel_processed.xlsx", index=False)
```
